```python
Ranked2 = pd.read_csv('datadump_s5-006.csv')
Ranked2.head()
```

```python
Ranked2 = Ranked2.dropna(how='any',axis=0)
```

```python
Ranked2['winrole']= Ranked2['winrole'].replace('Defender',0)
Ranked2['winrole']= Ranked2['winrole'].replace('Attacker',1)
Ranked2['winrole']= Ranked2['winrole'].astype(int)
```

```python
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('DefendersEliminated',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('AttackersSurrendered',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('DefendersEliminated',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('AttackersKilledHostage',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('TimeExpired',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('BombExploded',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('DefuserDeactivated',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('AllTeamsDead',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('BombDeactivated_OneBomb',1)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('ObjectiveProtected',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('NoEnemies',0)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('AttackersEliminated',1)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('HostageExtracted',1)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('ObjectiveCaptured',1)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('DefendersSurrendered',1)
Ranked2['endroundreason']= Ranked2['endroundreason'].replace('DefendersKilledHostage',1)

Ranked2['endroundreason']= Ranked2['endroundreason'].astype(int)
Ranked2.head()
```

```python
Ranked2_train = Ranked2.sample(frac=0.75,random_state=99)
Ranked2_test = Ranked2.loc[~Ranked2.index.isin(Ranked2_train.index),:]
```

```python
from sklearn import tree

features = Ranked2_train[['endroundreason','roundduration','roundnumber']]
labels = Ranked2_train['winrole']

clf = tree.DecisionTreeClassifier()
cls = clf.fit(features,labels)

test_features = Ranked2_test[['endroundreason','roundduration','roundnumber']]

predicted_values = clf.predict(test_features)
print(predicted_values)
```
